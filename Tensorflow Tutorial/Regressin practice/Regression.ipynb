```python
import csv
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
climate_change=pd.read_csv('climate_change.csv')
```

```python
climate_change=pd.DataFrame(climate_change)
```

```python
def add_layer(inputs,in_size,out_size,activation_function=None):
    Weights=tf.Variable(tf.random_normal([in_size,out_size]))
    biases=tf.Variable(tf.zeros([1,out_size])+0.1)
    Wx_plus_b=tf.matmul(inputs,Weights)+biases
    if activation_function is None:
        outputs=Wx_plus_b
    else:
        outputs=activation_function(Wx_plus_b)
    return outputs
```

```python
climate_change.tail()
```

```python

Co2=climate_change['CO2'][:,np.newaxis]
CH4=climate_change['CH4'][:,np.newaxis]

Temp=climate_change['Temp'][:,np.newaxis]

```

```python
Temp
```

```python
xs=tf.placeholder(tf.float32,[None,1])
ys=tf.placeholder(tf.float32,[None,1])
```

```python
l1=add_layer(xs,1,10,activation_function=tf.nn.relu)
prediction=add_layer(l1,10,1,activation_function=None)
loss=tf.reduce_mean(tf.reduce_sum(tf.square(ys-prediction),reduction_indices=[1]))
```

```python
train_step=tf.train.GradientDescentOptimizer(0.1).minimize(loss)
```

```python
init=tf.initialize_all_variables()
sess=tf.Session()
sess.run(init)
```

```python
for i in range(1000):
    sess.run(train_step,feed_dict={xs:Co2,ys:Temp})
    if i % 50==0:
        print(i,sess.run(loss,feed_dict={xs:Co2,ys:Temp}))
```
