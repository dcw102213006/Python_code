```python
##這支程式在把Amazon網站槌子商品的頁面的HTML爬下來存到mongo，是要解析槌子商品資訊的
from bs4 import BeautifulSoup
import requests
import re
import pymongo  
from pymongo import MongoClient
import pyodbc
import time
from datetime import datetime
import MongoConnect
headers = {
    'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1"
}  

# next_url='/knowledge/knowledge_lp.aspx?ArticleType=A&CategoryId=A&PageNumber=184&PageSize=10&dateS=&dateE='
```

```python
db=MongoConnect.mongoconnect()
one_document=db.Amazon_reviews_hammer_html
list=one_document.find({ '$and': [ { 'error': 0 }, { 'page': { '$gt': 274} } ] })
mongo_data=[]
for i in list:
    mongo_data.append(i)
    
```

```python
#每個row是一個槌子列表頁面，一個槌子列表頁面有多個槌子連結可以前往槌子資訊區
num=7161
for row in mongo_data:
    page=row['page']
    target_url=row['url']
    hammer_list_html=row['web_html']
    
    soup1=BeautifulSoup(hammer_list_html,'lxml')
    hammer_list=soup1.find_all('a',class_='a-link-normal a-text-normal')
    
    
    ##########hammer工具列表###################
    
    #for each 列表，前往該列表的每個槌子資訊區
    for hammer_href in hammer_list:
        if hammer_href['href'].startswith('https'):hammer_craw_url=hammer_href['href']
        if hammer_href['href'].startswith('/gp'):hammer_craw_url='https://www.amazon.com'+hammer_href['href']

        

        #####對每個hammer超連結爬該hammer的產品資訊的html:
        try:
            hammer_detail_html=requests.get(hammer_craw_url,headers = headers)  

            detail_soup=BeautifulSoup(hammer_detail_html.text,'lxml')
            
            

            

            
            bag={
                'num':str(num),
                'url':hammer_craw_url,
                'web_html':str(detail_soup),
                'error':0

            } 
            
            
        except Exception as e:
            bag={
               'num':'',
                'url':hammer_craw_url,
                'web_html':str(detail_soup),
                'error':1
            }
            #抓取失敗
            print(e)
            print(str(num)+".:Error find!Connect again...")
            time.sleep(20)
        finally:
            ###連接MONGO


        
            ##檢查Mongo看有沒有抓過該hammer的產品資訊的html
            one_document =db.Amazon_reviews_hammer_detail_html
            current_insert=one_document.find({'url':hammer_craw_url}).count()
            if(current_insert==0):
                one_document.insert_one(bag)
                print('page:'+str(page)+'\nhammer工具:'+hammer_craw_url)
                num=num+1
            if(current_insert>0):
                print('page:'+str(page)+'\nhammer工具:'+hammer_craw_url)
                print("此工具資訊區的html已抓過了")
            
            time.sleep(5)   
  

    
```
